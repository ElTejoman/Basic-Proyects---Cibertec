## Herencia
La herencia es la técnica por la cual una clase se crea a partir de otras de las que “hereda” sus atributos y métodos. Es característica de la POO (Programacion Orientada a Objetos) busca reutilizar el código. En Python, la relación de herencia se define pasando la clase padre como parámetro en la definición de la clase hija.


```python
class Automovil2:
  def __init__(self,marca, modelo, color,puertas,combustible):
    self.marca = marca
    self.modelo = modelo
    self.color = color
    self.puertas = puertas
    self.combustible = combustible
```

```python
class carros_peru(Automovil2):
  pass


au1 = carros_peru("Chevrolet","Onix","Rojo",5,"Gasolina")
print(au1.color)
```

```python
class carros_peru2(Automovil2):
  def __init__(self,marca, modelo, color,puertas,combustible,pais_importa):
    Automovil2.__init__(self,marca, modelo, color,puertas,combustible)

    self.pais_importa = pais_importa
```

```python
au2 = carros_peru2("Chevrolet","Onix","Rojo",5,"Gasolina","Perú")
print(au2.color)
print(au2.pais_importa)
```

```python
class Persona:# Clase base
    "Clase que representa una persona."
    def __init__(self, identificacion, nombre, apellido):
        "Constructor de Persona"
        self.identificacion = identificacion
        self.nombre = nombre
        self.apellido = apellido
```

```python
class AlumnoCIBERTEC(Persona):# clase hija
  "Clase que representa a un alumno de Cibertec."
  def __init__(self, identificacion, nombre, apellido, codigo_alumno):
    "Constructor de Cibertec"
    # llamamos al constructor de Persona
    Persona.__init__(self, identificacion, nombre, apellido)
    # agregamos el nuevo atributo
    self.codigo_alumno = codigo_alumno

  def __str__(self):
    "Devuelve una cadena representativa del alumno"
    return "Codigo: %s, Nombre:%s, Apellido: %s, DNI: %s" % (self.codigo_alumno,self.nombre, self.apellido, self.identificacion)
```

```python
a = AlumnoCIBERTEC("DNI 35123456", "Joaquin", "Sanchez", "20223654")
print(a)
```

```python
class AlumnoCIBERTEC2(Persona):
  "Clase que representa a un alumno de Cibertec."
  def __init__(self, identificacion, nombre, apellido, codigo_alumno):
    "Constructor de Cibertec"
    # llamamos al constructor de Persona
    Persona.__init__(self, identificacion, nombre, apellido)
    # agregamos el nuevo atributo
    self.codigo_alumno = codigo_alumno
```

```python
a = AlumnoCIBERTEC2("DNI 35123456", "Joaquin", "Sanchez", "20223654")
print(a.codigo_alumno)
print(a.nombre)
print(a.apellido)
print(a.identificacion)
```

## Polimorfismo
El concepto de polimorfismo hace referencia a la capacidad que puede tener un elemento de adoptar diferentes formas. Visto desde la perspectiva de las funciones, se refiere a los distintos comportamientos que puede tener una misma función en base a los tipos de datos operados.

```python
class Automovil3:
  def __init__(self,marca, modelo, color,puertas,combustible):
    self.__marca = marca # Atributos privados
    self.modelo = modelo
    self.color = color
    self.puertas = puertas
    self.combustible = combustible
    self.__estaEncendido = False # Atributos privados

  def encender(self): # creando un método
    self.__estaEncendido = True
    print("El auto está encendido")

  def apagar(self):
    self.__estaEncendido = False
    print("El auto está apagado")

  def get_encendido(self):# encender al carro
    return self.__estaEncendido

  def get_marca(self): # Obtener la marca del carro
    return self.__marca

  def set_marca(self, marca):# Modificar la marca
    self.__marca = marca
```

```python
auto1 = Automovil3("Chevrolet","Onix","Rojo",4,"Gasolina")
print(auto1.get_encendido())
```

```python
print(auto1.encender())
print(auto1.get_encendido())
```

```python
class Maquina:
  def __init__(self):
    self._estaEncendido = False

  def get_encendido(self):
    return self._estaEncendido


  def encender(self):
    self._estaEncendido = True
    print("La máquina está encendida")

  def apagar(self):
    self._estaEncendido = False
    print("La máquina está apagada")

```

```python
def iniciar(objeto):
  objeto.encender()

auto1 = Automovil3("Chevrolet","Onix","Rojo",4,"Gasolina")
maquina1 = Maquina()

iniciar(auto1)
iniciar(maquina1)
```
