## Aplicaciones del algebra matricial




Regrecion Lineal simple y muultiple

$$y=b_0+ b_1x_1+b_2x_2+b_3x_3$$

Funcion de regression usando notacion matricial:

$$\left[\begin{array}{cc}y_1\\y_2\\y_3\\.\\.\\y_n\end{array}\right]=\left[\begin{array}{cc}1 & x_{11} & x_{12} & x_{13} & .& .& x_{1p}\\1 & x_{21} & x_{22} & x_{23} & .& .& x_{2p} \\1 & x_{31} & x_{32} & x_{33} & .& .& x_{3p} \\. & . & . & . & .& .& .\\. & . & . & . & .& .& . \\1 & x_{n1} & x_{n2} & x_{n3} & .& .& x_{np}\end{array}\right]\cdot\left[\begin{array}{cc}\beta_0\\\beta_1\\\beta_2\\.\\.\\\beta_p\end{array}\right] + \left[\begin{array}{cc}\epsilon_1\\\epsilon_2\\\epsilon_3\\.\\.\\\epsilon_n\end{array}\right]$$

Donde:

$$\mathbf{X} = \left[\begin{array}{cc}1 & x_{11} & x_{12} & x_{13} & .& .& x_{1p}\\1 & x_{21} & x_{22} & x_{23} & .& .& x_{2p} \\1 & x_{31} & x_{32} & x_{33} & .& .& x_{3p} \\. & . & . & . & .& .& .\\. & . & . & . & .& .& . \\1 & x_{n1} & x_{n2} & x_{n3} & .& .& x_{np}\end{array}\right], \mathbf{y} = \left[\begin{array}{cc}y_1\\y_2\\y_3\\.\\.\\y_n\end{array}\right], \beta = \left[\begin{array}{cc}\beta_0\\\beta_1\\\beta_2\\.\\.\\\beta_p\end{array}\right], \epsilon = \left[\begin{array}{cc}\epsilon_1\\\epsilon_2\\\epsilon_3\\.\\.\\\epsilon_n\end{array}\right]$$

Ecuacion general de la regreción
$$\mathbf{y} = \mathbf{X}.\beta + \epsilon$$
Obtencion de los coeficientes

$$\hat{\mathbf{\beta}} = \left(\mathbf{X}^T \cdot \mathbf{X} \right)^{-1} \cdot \mathbf{X}^T \cdot \mathbf{y}$$

Donde
$\mathbf{X}^T$: Transpuesta de $\mathbf{X}^T$, $(^{-1})$: hace referencia a la inversa de la matriz, ($\cdot$) es el producto matricial

Siguiendo esta ultima formula se epude estimar los coeficientes

## Funciones de distribución de probabilidad

### Distribución Normal

```python
np.random.seed(2630)
#std, var, nanmean, nanstd, nanvar
#loc= promedio
#scale=sd=desviacion estandard
#size=n
norm_sim = np.random.normal(loc=15, scale=3, size=200)
print(norm_sim[0:10])
print(np.mean(norm_sim))#promedio
np.std(norm_sim)
```

```python
# hist, bins = np.histogram(norm_sim)
import seaborn as sns
from scipy.stats import norm

ax = sns.distplot(norm_sim,
                  #bins=50,
                  kde=False,
                  fit=norm,
                  hist_kws={"linewidth": 15,'alpha':1,"color":"#ffdfba"},
                  fit_kws={'color': 'green', 'linewidth': 2.5},
                  #kde_kws={'color': 'red', 'linewidth': 2.5}
                  )
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

plt.show()

```

Asumiendo que los datos tiene distribucion normal. ¿Cuál es la probabilidad de tener valores menores o iguales a 15?

```python
mean = np.mean(norm_sim)
sd= np.std(norm_sim)
prob = norm.cdf(x=10, loc=mean, scale=sd)
print("promedio: ", mean)
print("desviación estandard: ", sd)
print("Probabilidad X <= 15: ", prob)
```

### Distribución Poisson

```python
#lambda=promedio=media
pois_sim = np.random.poisson(lam=4, size=100)
print(pois_sim)
print(np.mean(pois_sim))
```

```python
a = np.hstack(pois_sim)
plt.hist(a, bins='auto')
```

```python
# Define the Poisson distribution parameter lambda
lam = np.mean(pois_sim)

# Create an array of x values
x = np.arange(0, 15).tolist()

# Create the Poisson probability mass function
pmf = poisson.pmf(x, lam)

# Create the plot
plt.plot(x, pmf, 'bo', ms=8)
plt.vlines(x, 0, pmf, colors='b', lw=5)
plt.title('Poisson Probability Mass Function')
plt.xlabel('Number of events')
plt.ylabel('Probability')
plt.show()
```

## Imagen digital como matriz numérica
