## Clases
Las clases son definiciones que representan entidades que tienen características y que pueden ejecutar operaciones. La materialización de un elemento perteneciente a una clase se conoce como objeto. Un objeto es una instancia de una clase y se construye en memoria según la definición de atributos (características) y métodos (operaciones) de la misma.

En el siguiente código estamos viendo dos aspectos importantes, estamos definiendo el atributo esta Encendido como privado mediante el uso  de "__", por otro lado para modificar este argumento se creo los **métodos** "encender" y "apagar". Finalmente para acceder a este argumento, se creo el **método** "get encendido.

Otro aspecto importante es el **encapsulamiento**, mediante el cual primero definimos atributos privados al que solo podemos modificar y acceder mediante funciones definidas (getter y setter respectivamente). En ele ejmplo podemos ver que podemos modificar la marca con la funcion set_marca() y acceder a la marca con la funcion get_marca()

```python

class Automovil3:
  def __init__(self,marca, modelo, color,puertas,combustible):
    self.__marca = marca # Atributos privados
    self.modelo = modelo
    self.color = color
    self.puertas = puertas
    self.combustible = combustible
    self.__estaEncendido = False # Atributos privados

  def area_circulo(self):
    return(self.puertas**2)


  def encender(self): # creando un método
    self.__estaEncendido = True

  def apagar(self):
    self.__estaEncendido = False

  def get_encendido(self):# encender al carro
    return self.__estaEncendido

  def get_marca(self): # Obtener la marca del carro
    return self.__marca

  def set_marca(self, marca):# Modificar la marca
    self.__marca = marca
```

```python
auto1 = Automovil3("Chevrolet","Onix","Rojo",4,"Gasolina")
print(auto1.get_encendido())

auto1.encender()
print(auto1.get_encendido())

auto1.apagar()
print(auto1.get_encendido())
```

```python
class Automovil4:
  def __init__(self,marca, modelo, color,puertas,combustible):
    self.__marca = marca
    self.modelo = modelo
    self.color = color
    self.__puertas = puertas
    self.combustible = combustible


  def get_puertas(self):
    return self.__puertas

  def set_puertas(self,npuertas):
    if npuertas<=1 or npuertas>5:
      raise ValueError("Advertencia!: Las puertas deben estar entre 2 y 5")
    self.__puertas = npuertas
```

```python
autom1 = Automovil4("Chevrolet","Onix","Rojo",8,"Gasolina")
print(autom1.get_puertas())
```

```python
autom1.set_puertas(2)
print(autom1.get_puertas())
```
